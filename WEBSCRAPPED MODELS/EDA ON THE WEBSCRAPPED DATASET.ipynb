# IMPORTING ALL THE LIBRARIES.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOADING THE DATASET.

```python
data=pd.read_excel(r"C:\Users\pc\OneDrive\Desktop\DATASET_SCRAPPED_FROM_IMDb..xlsx")
```

```python
data.head()
```

# CHECKING FOR NAN VALUES AND TOTAL NUMBER OF VALUES IN DATASET.

```python
data.info()
```

# REMOVING THE COLUMN1 AS IT IS SAME AS THE INDEX OF THE DATAPOINT.

```python
data.drop('Column1',axis=1)
```

# CONVERTING THE 'time_duration' of a movie to NUEMRICAL FORMAT TO ANALIZE IN THE SCATTER PLOT.

```python
movie_time=[]
for i in data['time_duration']:
    movie_time.append(int(i[0:len(i)-3].lstrip()))
```

```python
data['movie_time']=movie_time  # ADDING A NEW FEATURE IN THE DATASET.
```

```python
data.head()
```

# UNIVARIATE ANALYSIS.

1. HOW USERS GIVE RATINGS.

```python
sns.countplot(x=data['rating'])
```

SO FROM THE GRAPH WE CAN SAY THAT THERE ARE ONLY FEW USER WHICH GIVE VERY HIGH RATING TO MOVIES OR ONLY FEW MOVIES ARE THERE WHCIH RECIEVES VERY HIGH RATINGS.

2. HOW CRITICS GIVE RATINGS.

```python
sns.countplot(x=data['metascore'])
plt.xticks(rotation=90)
print()
```

AS COMPARED TO USERS CRITICS RARINGS ARE ALSO THERE IN THE HIGH VALUES LIKE IN THE RANGE OF 90 TO 98.

# BIVAIRATE ANALYSIS.

1. PLOTTING THE PAIR PLOT AND ANALYSING HOW DOES EACH FEATURE IS DEPENDING ON THE OTHER FEATURES IN THE DATASET.

```python
sns.pairplot(data[['rating','year','metascore','movie_time']])
```

# FINDING THE CORELATION BETWEEN THE FEATURES USING CORELATION MATRIX WHICH HELPS US TO FIND THE COREALTION BETWEEN FEATURES WITH THE HELP OF NUMBERS.

```python
corelational_matrix=data[['year','rating','metascore','movie_time']].corr()
corelational_matrix
```

```python
sns.heatmap(corelational_matrix,annot=True,linewidths=0.9,cmap='crest')
```
