# CREATING BAG OF WORDS.

```python
vectors=np.zeros((data.shape[0],len(vocabulary_)))
for i in range(data.shape[0]):
    for j in range(0,len(vocabulary_)):
        if vocabulary_[j] in data.iloc[i].model_features:
            vectors[i][j]=1
        else:
            vectors[i][j]=0
```
