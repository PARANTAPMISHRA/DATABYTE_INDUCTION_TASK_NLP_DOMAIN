# CREATING USER - ITEM MATRIX FOR PERFORMING MATRIX FATORIZATION WHICH IS COLABORATIVE BASED MOVIE RECOMMENDATION SYSTEM. 

FINDING THE UNIQUE MOVIE ID AND UNIQUE ITME ID TO CREATE THE USER-ITEM-MATRIX.

```python
unique_user_ids=len(data['userId'].unique())
unique_movie_ids=len(data['movieId'].unique())
user_item_matrix=np.zeros((unique_user_ids,unique_movie_ids))
```

```python
data[data['movieId']==1]
```

CREATING USER ITEM MATRIX.

```python
for i in range(0,len(data['movieId'].unique())):
    user_voted=np.array(data[data['movieId']==data['movieId'].unique()[i]].userId)
    user_votes=np.array(data[data['movieId']==data['movieId'].unique()[i]].rating)
    for j in range(0,len(data['userId'].unique())):
        if data['userId'].unique()[j] in user_voted:
            rating=user_votes[list(user_voted).index(data['userId'].unique()[j])]
            user_item_matrix[j][i]=rating
        else:
            user_item_matrix[j][i]=0
            
```

# CREATING A CLASS TO COMPUTE THE MATRIX FACTORIZATION.

```python
class Matrix_Factorization():
    def __init__(self,epochs=10,features=3,learning_rate=0.001):
        # INITIALIZING THE USER AND ITEMS FEATUES WITH RANDOM VALUES.
        self.user_features=np.random.rand(user_item_matrix.shape[0],features)
        self.item_features=np.random.rand(user_item_matrix.shape[1],features).T
        self.epochs=epochs
        self.learning_rate=learning_rate
        self.features=features
    def _compute_MF(self,y_true):
        for i in range(self.epochs):
            for j in range(y_true.shape[0]):
                for k in range(y_true.shape[1]):
                    if y_true[j][k]>0:
                        loss=y_true[j][k]-np.dot(self.user_features[j,:],self.item_features[:,k])
                        for l in range(self.features):
                            self.user_features[j][l]=self.user_features[j][l]+(2*self.learning_rate*loss*self.item_features[l][k])
                            self.item_features[l][k]=self.item_features[l][k]+(2*self.learning_rate*loss*self.user_features[j][l])
            print('done')
        return(np.dot(self.user_features,self.item_features))
```
