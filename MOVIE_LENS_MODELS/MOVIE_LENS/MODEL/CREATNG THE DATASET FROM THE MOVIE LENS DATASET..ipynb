```python
movieId_real=[]   # LIST TO STORE THE MOVIE_ID WHICH IS USED TO GET THE TMDbID FROM THE LINK DATASET.
title=[]          # TO STORE THE NAME OF THE MOVIE.
genre=[]          # TO STROE THE GENRE OF THE MOVIE.
for _,i in enumerate(np.array(movie)):
    title.append(i[1])
    genre.append(i[2])
    movieId_real.append(i[0])
```

# CREATING A DICTIONARY TO STORE THE RATING OF A MOVIE AS KEY AND AVERAGE RATING AS THE VALUE.

```python
d={}
for i in movieId_real:
    d[i]=[]
```

```python
rating_values=[]
[rating_values.append(i[1]) for i in(np.array(rating))]
average_rating=int(np.sum(rating_values)/len(rating_values))
print(average_rating)
```

```python

for i in movieId_real:

    for _,j in enumerate(np.array(rating)):
            if i==j[0]:
                d[i].append(j[1])
    if len(d[i])==0:   
        d[i]=average_rating    # IF THER ARE NO RATINGS FOR THE GIVEN MOVIES THEN ASSIGNING AVERAGE RATNG OF ALL THE RATINGS IN THE RATINGS DATA SET.
    else:
        d[i]=sum(d[i])/len(d[i])  #  IF RATINGS ARE PRESENT THEN ASSIGNING AVERAGE RATING OF THE PROVIDED RATINGS FOR THE GIVEN MOVIE.
        
    
```

```python
Rating=list(d.values())
```

```python
Year=[]
Title=[]
for i in title:
    l=len(i)
    year=i[l-5:l-1:1]
    t=i[0:l-7:1]
    Year.append(year)
    Title.append(t) 
```

# CREATING THE DATAFRAME OUT OF THE FOLLOWING FEATURES.

```python
data=pd.DataFrame({'movie_id':movieId_real,'title':Title,'year':Year,'genre':genre,'rating':Rating})
```
